## Reducing the digits in the gross and budget column

```python
movies["Gross"]=movies["Gross"]/1000000
movies["budget"]=movies["budget"]/1000000
```

```python
movies[["Gross","budget"]].head(10)
```

```python
movies.describe()
```

```python
movies["profit"]=movies["Gross"]-movies["budget"]
movies["profit"].head()
```

```python
movies["profit"].describe()
```

```python
#identify movie loss of $96 millions
movies[movies["profit"]==min(movies["profit"])]
#or
movies[movies["profit"]==-96.179906]
```

```python
#identify movie with max profit
movies[movies["profit"]==max(movies["profit"])]
```

```python
df_g1=df_genre.groupby('genre_1').sum()
```

```python
df_genre[df_genre['genre_1']=='Action']
```

```python
df_g1
```

```python
df_g2=df_genre.groupby('genre_2').sum()
```

```python
df_g3=df_genre.groupby('genre_3').sum()
```

```python
df_g3
df_g2
```

```python
#add the grouped data and store it in a new dataframe
df_add=df_g1.add(df_g2,fill_value=0).add(df_g3,fill_value=0)
```

```python
df_add.head()
df_add.shape
```

```python
df_g1.cnt
df_g2.cnt
df_g3.cnt
```

```python
#extracting genres with atleast cnt >10
top10_genres= df_add[df_add['cnt']>=10]
```

```python
top10_genres.head()
```

```python
top10_genres.sort_values('cnt',ascending=False)[:5]
```
